```python
def getTicker(gsid):
    import requests
    import json

    client_id = r'07597d01c4a647278eee197b16f25b90'
    client_secret = r'b6eb791792918051097310c26953b80c6afd65e3ad945a50c1360d041bb3446b'

    auth_data = {
        'grant_type'    : 'client_credentials',
        'client_id'     : client_id,
        'client_secret' : client_secret,
        'scope'         : 'read_content read_financial_data read_product_data read_user_profile'
    }

    # create session instance
    session = requests.Session()

    # make a POST to retrieve access_token
    auth_request = session.post('https://idfs.gs.com/as/token.oauth2', data = auth_data)
    access_token_dict = json.loads(auth_request.text)
    access_token = access_token_dict['access_token']

    # update session headers
    session.headers.update({'Authorization':'Bearer '+ access_token})

    # test API connectivity
    request_url = 'https://api.marquee.gs.com/v1/assets/data/query'
    request_query = {"where": {"gsid":[str(gsid)]}, "fields": ["ticker"]}
    request = session.post(url=request_url, json=request_query)
    
    results = json.loads(request.text)
    
    return(results["results"][0]["ticker"])
    
biglist = pd.read_json('output.json')

def checkBigList(x):
    secure = pd.DataFrame({'gsid':biglist['gsid'], 'date':biglist['date'], 'integratedScore':biglist['integratedScore']})
    secure = secure.set_index('gsid')
    secure = secure.loc[x]
    
    dictionary = {}
    dictionary['Ticker'] = str(getTicker(x))
    
    try:
        mean1 = secure.set_index('date').loc['2012'].mean()
        dictionary['2012'] = mean1[0]
    except:
        mean1 = np.nan
        dictionary['2012'] = mean1

    
    mean2 = secure.set_index('date').loc['2013'].mean()
    dictionary['2013'] = mean2[0]
    mean3 = secure.set_index('date').loc['2014'].mean()
    dictionary['2014'] = mean3[0]
    mean4 = secure.set_index('date').loc['2015'].mean()
    dictionary['2015'] = mean4[0]
    mean5 = secure.set_index('date').loc['2016'].mean()
    dictionary['2016'] = mean5[0]
    
    try:
        mean6 = secure.set_index('date').loc['2017'].mean()
        dictionary['2017'] = mean6[0]
    except:
        mean6 = np.nan
        dictionary['2017'] = mean6
        
    return(dictionary)
    
dataset = []    
y = 0
z = 0
for x in biglist['gsid']:
    if (y != x):
        y = x
        dataset.append(checkBigList(x))

dataframe = pd.DataFrame(dataset)
dataframe
```
